```python
import pandas as pd
```

```python
# import data from Definiteve Healthcare
hs_beds = pd.read_csv("Definitive_Healthcare__USA_Hospital_Beds.csv")
```

```python
hs_beds.head()
```

```python
# look for California data only
ca_beds = hs_beds[hs_beds['HQ_STATE'] == 'CA']
```

```python
# aggregate the records in each county using sum
county_level = ca_beds.groupby(['COUNTY_NAME']).sum()
```

```python
# create a CA copy to aggregate BED_UTILIZATION and AVG_VENTILATOR_USAGE using mean
temp = ca_beds.copy()[['BED_UTILIZATION','COUNTY_NAME','AVG_VENTILATOR_USAGE']].sort_values(by='COUNTY_NAME').reset_index()
county_level_tmp = temp.groupby(['COUNTY_NAME']).mean()
```

```python
county_level_tmp
```

```python
# drop the BED_UTILIZATION and AVG_VENTILATOR_USAGE from sum aggregation and join the mean aggregations
county_level = county_level.drop(['BED_UTILIZATION','AVG_VENTILATOR_USAGE'], axis=1)
ca_county = county_level.join(county_level_tmp,on='COUNTY_NAME')
```

```python
ca_county
```

```python
# Cleanup the columns
ca_county = ca_county.drop(['X', 'Y', 'OBJECTID', 'HQ_ZIP_CODE', 'STATE_FIPS', 'CNTY_FIPS', 'FIPS', 'index'],axis=1)
```

```python
ca_county
```

```python
# save it to .csv for visualizations in Tableau and modeling steps
ca_county.to_csv("CA_COUNTY_BEDS.csv")
```
